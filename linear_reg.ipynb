```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = 2
c = 3

X = np.random.rand(200)
X
```

```python
ya = (m*X +c)  + np.random.rand(200)
ya
```

```python
# MSE = Sum((ya-yp)^2)/n
# dm = -2/n sum(X(ya-yp))
# dc = -2/n sum(ya-yp)
```

```python
yp = m*X +c 
n = 200
```

```python
sum(X*(ya-yp))
```

```python

m_d = -(2/n) * sum(X*(ya-yp)) 
c_d = -(2/n) * sum(ya-yp)
```

```python
# m_new = m- lr* dmse/dm 
# c_new = c - lr* dmse/dm
```

```python
m = 2
c =3 
n = len(X)

loss = []
slope = []
intercepts = []

epoch = 1000
lr = 0.001

for i in range(epoch):
   yp = m*X + c

   mse = np.mean(np.square(ya-yp))
   loss.append(mse)

   m_d = -(2/n) * sum(X*(ya-yp)) 
   c_d = -(2/n) * sum(ya-yp)
   
   m = m - lr*m_d
   c = c - lr*c_d

   slope.append(m)
   intercepts.append(c)

   if i%100==0:
      print(f"Slope {m} Intercept {c}")

   

```
